```python
import numpy as np
from scipy import stats
from scipy import signal
from scipy.fftpack import fft2
from scipy import linalg
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

t = np.array([0, 1, 2, 3, 4, 5])
v = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def model(t, a, b, c):
    return a*t**2 + b*t + c

popt, _ = curve_fit(model, t, v)
a, b, c = popt

t_fit = np.linspace(0, 5, 100)
v_fit = model(t_fit, a, b, c)

print("Curve Fit Parameters: a=%.3f, b=%.3f, c=%.3f" % (a, b, c))

plt.plot(t, v, 'o', label="Data")
plt.plot(t_fit, v_fit, '-', label="Fitted Curve")
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats
from scipy import signal
from scipy.fftpack import fft2
from scipy import linalg
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


t = np.array([0, 1, 2, 3, 4, 5])
v = np.array([2, 3.1, 7.9, 18.2, 34.3, 56.2])

def model(t, a, b, c):
    return a*t**2 + b*t + c

popt, _ = curve_fit(model, t, v)
a, b, c = popt

t_fit = np.linspace(0, 5, 200)
v_fit = model(t_fit, a, b, c)

plt.figure(figsize=(7,5))
plt.plot(t, v, 'o', label="Original Data", markersize=8)
plt.plot(t_fit, v_fit, '-', label="Fitted Quadratic Curve", linewidth=2)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity vs Time with Quadratic Curve Fit")
plt.legend()
plt.grid(True)
plt.show()

print("Fitted Parameters: a=%.3f, b=%.3f, c=%.3f" % (a, b, c))
```
